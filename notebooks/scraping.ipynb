```python
from google_play_scraper import reviews, Sort
import pandas as pd
import os
```

```python
APPS = {
    "Commercial Bank of Ethiopia": "com.combanketh.mobilebanking",
    "Bank of Abyssinia": "com.boa.boaMobileBanking",
    "Dashen Bank": "com.dashen.amoleplus"
}
```

```python
all_reviews = []

for bank_name, app_id in APPS.items():
    print(f"Scraping {bank_name}...")
    result, _ = reviews(
        app_id,
        lang='en',
        country='et',
        sort=Sort.NEWEST,
        count=500
    )
    df = pd.DataFrame(result)
    df['bank'] = bank_name
    df['source'] = 'Google Play'
    print(f"Got {len(df)} reviews for {bank_name}")
    all_reviews.append(df)
```

```python
combined = pd.concat(all_reviews, ignore_index=True)
combined = combined[['reviewId','content','score','at','bank','source']]
combined.columns = ['id','review','rating','date','bank','source']
combined.drop_duplicates(subset='id', inplace=True)
combined.dropna(subset=['review','rating'], inplace=True)
combined['date'] = pd.to_datetime(combined['date']).dt.strftime('%Y-%m-%d')
os.makedirs('data/raw', exist_ok=True)
combined.to_csv('data/raw/reviews.csv', index=False)
print(f"Total reviews: {len(combined)}")
combined.head()
```
